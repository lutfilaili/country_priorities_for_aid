# IMPORTING LIBRARIES

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

# DATA PREPROCESSING

```python
# collect dataset
df = pd.read_csv('https://raw.githubusercontent.com/smirzap/AI-dengan-DL/main/data/Country-data.csv')
df
```

```python
# Improting Library for Visualization
from matplotlib import pyplot as plt
import plotly.express as px

# Create an interactive scatter plot
fig = px.scatter(df, x='imports', y='income',
                 title='Plotting',
                 labels={'imports': 'Imports', 'income': 'Income'})

# show the plot
fig.show()
```

```python
# check the data generally and transpose them
df.describe().T
```

```python
# remove country column
df1 = df.drop('country', axis=1)
```

```python
df1.head()
```

```python
# standardize the column value
scaler = StandardScaler()
dataset_std = scaler.fit_transform(df1)
print(dataset_std)
```

# MODELING - KMEANS

```python
# KMeans algorithm
model = KMeans(2)
```

```python
# Clustering with KMeans for 2 Clusters
clusters = model.fit_predict(dataset_std)
```

```python
clusters
```

```python
# adding new cluster (country) to df
df['cluster'] = clusters
```

```python
df.head(15)
```

```python
# Create an interactive scatter plot
fig = px.scatter(df, x='income', y='imports', color='cluster',
                 hover_data=['country'],
                 title='Clustered Countries',
                 labels={'income': 'Income', 'child_mort': 'Child Mortality'})

# Show the plot
fig.show()
```
